```python
# Block 2: Load the Model (The Brain)
from transformers import pipeline

print("Waking up the AI... (This takes about 30 seconds)")
# We use 'text2text-generation' because we give it text, and it gives us text back.
agent_brain = pipeline("text-generation", model="google/flan-t5-large", device=0)

print("Agent is awake! 🤖")
```

```python
# Block 3: Define the Tools
def calculator_tool(expression):
    """Solves math problems."""
    try:
        # This turns text like "55 * 10" into the number 550
        return eval(expression)
    except:
        return "Error: I couldn't calculate that."

print("Calculator tool is ready! 🧮")
```

```python
# Block 4: The Agent System
def run_agent(user_input):
    # 1. THE SYSTEM PROMPT (Intent Engineering)
    # We tell the AI clearly what its job is.
    system_prompt = (
        "You are a helpful AI Agent. "
        "If the user asks for a math calculation, answer with 'CALCULATE: ' followed by the equation. "
        "If the user asks a normal question, just answer it normally.\n\n"
        f"User: {user_input}\n"
        "Agent:"
    )

    # 2. THE BRAIN THINKS
    response = agent_brain(system_prompt, max_length=100)[0]['generated_text']

    # 3. THE LOGIC (Reasoning)
    if "CALCULATE:" in response:
        # The agent decided to use the tool!
        math_equation = response.replace("CALCULATE:", "").strip()
        print(f"🤖 (Thinking): I need to use my calculator for '{math_equation}'...")

        # Use the Tool
        result = calculator_tool(math_equation)

        # Final Answer
        return f"The answer is {result}."
    else:
        # The agent decided it doesn't need tools
        return response

print("Agent Logic is loaded! 🧠")
```

```python
# Block 5: Test the Agent

# Test 1: A normal question (No tools needed)
print("User: What is the capital of France?")
print("Agent:", run_agent("What is the capital of France?"))

print("-" * 20)

# Test 2: A math question (Tool needed!)
print("User: What is 25 * 4 + 10?")
print("Agent:", run_agent("What is 25 * 4 + 10?"))
```
