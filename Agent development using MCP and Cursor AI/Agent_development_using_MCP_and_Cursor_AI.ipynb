```python
%%writefile app.py
import streamlit as st
import yfinance as yf
import pandas as pd

# 1. SETUP THE PAGE
st.set_page_config(page_title="Global Finance Agent", page_icon="üí∞")
st.title("üí∞ AI Financial Agent")
st.markdown("Select a country to see its **Currency**, **Stock Market**, and **Headquarters Location**.")

# 2. THE BRAIN (DATA DICTIONARY)
# We map countries to their specific codes manually for accuracy
country_data = {
    "Japan": {
        "currency": "JPY",
        "ticker": "^N225", # Nikkei 225
        "exchange_name": "Tokyo Stock Exchange",
        "coords": [35.6812, 139.7745] # Tokyo Latitude/Longitude
    },
    "India": {
        "currency": "INR",
        "ticker": "^BSESN", # Sensex
        "exchange_name": "Bombay Stock Exchange (BSE)",
        "coords": [18.9291, 72.8339]
    },
    "USA": {
        "currency": "USD",
        "ticker": "^GSPC", # S&P 500
        "exchange_name": "New York Stock Exchange (NYSE)",
        "coords": [40.7069, -74.0113]
    },
    "UK": {
        "currency": "GBP",
        "ticker": "^FTSE", # FTSE 100
        "exchange_name": "London Stock Exchange",
        "coords": [51.5150, -0.0982]
    }
}

# 3. SIDEBAR SELECTION
selected_country = st.sidebar.selectbox("Choose a Country", list(country_data.keys()))
data = country_data[selected_country]

# 4. GET REAL-TIME DATA
st.header(f"üáØüáµ Analysis for {selected_country}")

# Section A: Currency
st.subheader(f"üí± Official Currency: {data['currency']}")
col1, col2, col3 = st.columns(3)

# Function to get exchange rate
def get_rate(from_curr, to_curr):
    if from_curr == to_curr: return 1.0
    ticker = f"{from_curr}{to_curr}=X"
    try:
        return yf.Ticker(ticker).history(period="1d")['Close'].iloc[-1]
    except:
        return 0.0

usd_rate = get_rate(data['currency'], "USD")
eur_rate = get_rate(data['currency'], "EUR")
inr_rate = get_rate(data['currency'], "INR")

col1.metric(label="To USD ($)", value=f"{usd_rate:.4f}")
col2.metric(label="To EUR (‚Ç¨)", value=f"{eur_rate:.4f}")
col3.metric(label="To INR (‚Çπ)", value=f"{inr_rate:.4f}")

# Section B: Stock Market
st.subheader(f"üìà Major Index: {data['ticker']}")
stock = yf.Ticker(data['ticker'])
hist = stock.history(period="1mo")

# Show the latest price
current_price = hist['Close'].iloc[-1]
st.metric(label="Current Value", value=f"{current_price:.2f}")

# Draw the Chart
st.line_chart(hist['Close'])

# Section C: Google Maps Location
st.subheader(f"üìç HQ: {data['exchange_name']}")
map_data = pd.DataFrame({'lat': [data['coords'][0]], 'lon': [data['coords'][1]]})
st.map(map_data, zoom=12)
```
