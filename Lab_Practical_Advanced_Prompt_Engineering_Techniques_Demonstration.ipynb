```python
import os
from openai import OpenAI

# ⚠️ Paste your Groq key below inside quotes
GROQ_API_KEY = "YOUR_API_KEY"

client = OpenAI(
    api_key=GROQ_API_KEY,
    base_url="https://api.groq.com/openai/v1"
)

print("Groq API Connected Successfully")
```

```python
def run_prompt(prompt, model="llama-3.1-8b-instant"):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    return response.choices[0].message.content
```

```python
persona_prompt = """
You are a senior AI professor teaching engineering students.
Explain Machine Learning in simple technical terms.
"""

print("---- Persona Prompt ----")
print(run_prompt(persona_prompt, model="llama-3.1-8b-instant"))
```

```python
cognitive_prompt = """
Explain what SQL Injection is.
After explaining, verify your answer step-by-step and check if any important detail is missing.
"""

print("---- Cognitive Verifier ----")
print(run_prompt(cognitive_prompt))
```

```python
question_refine_prompt = """
Refine this vague question into a clear academic question, then answer it.

Question: "How does hacking work?"
"""

print("---- Question Refinement ----")
print(run_prompt(question_refine_prompt))
```

```python
info_prompt = """
User says: "My laptop is running very slow."

Provide possible technical reasons.
Then ask 3 diagnostic questions.
"""

print("---- Provide Info + Ask Questions ----")
print(run_prompt(info_prompt))
```

```python
root_prompt = """
You are an academic assistant.

Rules:
- Use headings
- Use bullet points
- Keep answers structured
- Be concise

Explain Artificial Intelligence.
"""

print("---- Root Prompt ----")
print(run_prompt(root_prompt))
```

```python
cot_prompt = """
If a student studies 3 hours per day for 4 days,
how many total hours?

Think step by step before giving the final answer.
"""

print("---- Chain of Thought ----")
print(run_prompt(cot_prompt))
```

```python
table_prompt = """
Compare Artificial Intelligence, Machine Learning,
and Deep Learning in tabular format.
"""

print("---- Tabular Format ----")
print(run_prompt(table_prompt))
```

```python
fill_prompt = """
Fill in the blanks:

1. CPU stands for ______.
2. AI means ______.
3. Phishing is a type of ______ attack.
"""

print("---- Fill in the Blank ----")
print(run_prompt(fill_prompt))
```

```python
rgc_prompt = """
Role: You are a cybersecurity expert.
Goal: Explain how to prevent phishing attacks.
Context: The explanation is for first-year students.

Provide structured explanation.
"""

print("---- RGC Pattern ----")
print(run_prompt(rgc_prompt))
```

```python
zero_shot_prompt = """
Classify the sentiment as Positive, Negative, or Neutral:

"I am extremely unhappy with this service."
"""

print("---- Zero Shot ----")
print(run_prompt(zero_shot_prompt))
```

```python
one_shot_prompt = """
Example:
Text: "I love this product."
Sentiment: Positive

Now classify:
Text: "This is the worst experience ever."
Sentiment:
"""

print("---- One Shot ----")
print(run_prompt(one_shot_prompt))
```

```python
few_shot_prompt = """
Text: "Amazing performance!"
Sentiment: Positive

Text: "Very bad service."
Sentiment: Negative

Text: "It was okay."
Sentiment: Neutral

Now classify:
Text: "Absolutely fantastic experience!"
Sentiment:
"""

print("---- Few Shot ----")
print(run_prompt(few_shot_prompt))
```
