```python
import os
from groq import Groq

GROQ_API_KEY = "YOUR_API_KEY"

client = Groq(api_key=GROQ_API_KEY)

# A simple helper function to make running prompts easier
def ask_groq(prompt, model="llama-3.3-70b-versatile"):
    completion = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.7,
        max_tokens=1024,
    )
    return completion.choices[0].message.content

print("Setup Complete! You are ready to prompt.")
```

```python
prompt_1 = """
[Persona]: You are a Senior Cyber Security Analyst with 15 years of experience in network defense.

[Root Prompt]: Explain the concept of a 'Zero Trust Architecture' to a new junior developer.

[Cognitive Verifier]: Before giving the explanation, list 3 fundamental questions you must ask yourself to ensure your explanation covers the core principles of Zero Trust properly. Then, answer them.
"""

response = ask_groq(prompt_1)
print(response)
```

```python
prompt_2 = """
[Instruction]: I want to prompt you to write a Python script for data analysis, but my prompt is vague.

[My Vague Prompt]: "Write a code to analyze data."

[Task]:
1. Critique my vague prompt.
2. Suggest 2 refined, specific versions of this prompt (one for Finance, one for Health).
3. Ask me which version I would like to proceed with.
"""

response = ask_groq(prompt_2)
print(response)
```

```python
prompt_3 = """
[Context / New Information]:
"Quantum computing uses qubits instead of bits. Unlike bits which are 0 or 1, qubits can exist in a state of superposition. This allows quantum computers to solve specific complex problems exponentially faster than classical computers."

[Task]:
Based on the text above, ask me (the user) 2 questions to check if I understood the core difference between a bit and a qubit.
"""

response = ask_groq(prompt_3)
print(response)
```

```python
prompt_4 = """
[Role]: You are an expert Nutritionist and Dietitian.
[Goal]: Create a 1-day meal plan formatted as a Markdown table.
[Context]: The client is a vegetarian athlete preparing for a marathon.

Output Format:
Please use a Markdown Table with columns: | Meal Type | Food Items | Protein Content (approx) |
"""

response = ask_groq(prompt_4)
print(response)
```
